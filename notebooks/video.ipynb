```python
import cv2
import time
from ultralytics import YOLO

print(f"\n{'='*20} ГЕНЕРАЦІЯ ФІНАЛЬНОГО ВІДЕО {'='*20}\n")

final_model_path = '/home/obahlai/Documents/detection_segmentation/runs/segment/yolov8n_seg_phase2_tuned/weights/best.pt'
model = YOLO(final_model_path)

input_video_path = '/home/obahlai/Documents/detection_segmentation/test/test21.MP4'  
output_video_path = '/home/obahlai/Documents/detection_segmentation/diploma_final_demo.mp4'

cap = cv2.VideoCapture(input_video_path)

if not cap.isOpened():
    print("❌ Помилка: Не вдалося знайти або відкрити test_video.mp4")
else:
    w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    out = cv2.VideoWriter(output_video_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, (w, h))

    print(f"Починаємо обробку {total_frames} кадрів...")
    frame_count = 0

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        start_time = time.time()
        
        results = model.predict(frame, imgsz=640, conf=0.3, verbose=False)

        for key in results[0].names.keys():
            results[0].names[key] = '' 

        if results[0].masks is not None:
            annotated_frame = results[0].plot(conf=True, labels=True, boxes=True, line_width=2)
        else:
            annotated_frame = frame.copy()

        process_time = time.time() - start_time
        curr_fps = 1.0 / process_time if process_time > 0 else 0

        cv2.putText(annotated_frame, "MILITARY VEHICLE DETECTION", (30, 50), 
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        cv2.putText(annotated_frame, f"IFPS: {curr_fps:.1f}", (30, 90), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)

        out.write(annotated_frame)
        frame_count += 1
        
        if frame_count % 50 == 0:
            print(f"⏳ Оброблено {frame_count}/{total_frames} кадрів...")

    cap.release()
    out.release()
    print(f"\n✅ Готово! Відео успішно збережено як:\n{output_video_path}")
```
