```python
import os
import glob

dataset_path = '/home/obahlai/Documents/detection_segmentation/Mil_Vechicle_Segmentation-1'

print(f"Починаємо очистку датасету: {dataset_path}")

txt_files = glob.glob(os.path.join(dataset_path, "**", "*.txt"), recursive=True)
cleaned_count = 0

for txt_file in txt_files:
    with open(txt_file, 'r') as f:
        lines = f.readlines()
    
    new_lines = []
    file_changed = False
    
    for line in lines:
        parts = line.strip().split()
        if len(parts) > 5:
            new_lines.append(line)
        else:
            file_changed = True
            
    if file_changed:
        with open(txt_file, 'w') as f:
            f.writelines(new_lines)
        cleaned_count += 1

print(f"✅ Готово! Очищено файлів: {cleaned_count}")
print("Тепер можна запускати тренування model.train(...)")
```
